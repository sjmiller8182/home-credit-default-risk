```python
import pandas as pd
import prince # can only install through the pipeline ( run <pip install prince> command )
import pprint
```

```python
famd = prince.FAMD(
    normalize=True # consider taking this out. especially if the code isn't running.
    , n_components=25 # the number of components that are computed. You only need two if your intention is to make a chart
    , n_iter=10 # the number of iterations used for computing the SVD
    , copy=True # if False then the computations will be done inplace
    , check_input=True
    , engine='auto' # what SVD engine to use (should be one of ['auto', 'fbpca', 'sklearn'])
    , random_state=42 # controls the randomness of the SVD results
    )
```

```python
#Fit FAMD object to data
famd = famd.fit(df.drop('TARGET', axis='columns')) # drop the target (response) variable. also, axis=1 is for first row
```

```python
pp = pprint.PrettyPrinter()
```

```python
import prince
pp.pprint(famd.explained_intertia_)
```

```python
# reference:
## https://pypi.org/project/prince/#factor-analysis-of-mixed-data-famd
## https://nextjournal.com/pc-methods/calculate-pc-mixed-data?change-id=CWQNw1kVRgQMFFzobMC2bo&node-id=d4243af6-f940-41fc-8ffa-a235bc135601
ax = famd.plot_row_coordinates(
    df
    , ax=None
    , figsize=(6, 6)
    , x_component=0
    , y_component=1
    , labels=df.index
    , color_labels=['TARGET {}'.format(t) for t in df['TARGET']]
    , ellipse_outline=False
    , ellipse_fill=True
    , show_points=True
 )
```
